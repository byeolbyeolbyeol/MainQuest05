```python
# 결측치 도메인 기반 처리
cols_fill_none = ["PoolQC", "MiscFeature", "Alley", "Fence", "FireplaceQu",
  "GarageType", "GarageFinish", "GarageQual", "GarageCond",
  "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2",
  "MasVnrType"]
for col in cols_fill_none:
  train[col] = train[col].fillna("None")
  test[col] = test[col].fillna("None")
```

```python
# LotFrontage → 동네별 중앙값으로 대체
grouped_median = train.groupby("Neighborhood")["LotFrontage"].median()
train["LotFrontage"] = train.apply(lambda row: grouped_median[row["Neighborhood"]] if pd.isnull(row["LotFrontage"]) else row["LotFrontage"], axis=1)
test["LotFrontage"] = test.apply(lambda row: grouped_median[row["Neighborhood"]] if pd.isnull(row["LotFrontage"]) else row["LotFrontage"], axis=1)
```

```python
# 나머지 결측치 채우기 (수치형 → 0, 범주형 → 최빈값)
for df in [train, test]:
  for col in df.columns:
    if df[col].dtype == "object":
      df[col] = df[col].fillna(df[col].mode()[0])
    else:
      df[col] = df[col].fillna(0)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# 파생변수 생성(총 생활면적, 총 욕실 수, 건축 후 경과 연도)
for df in [train, test]:
  df['TotalSF'] = df['1stFlrSF'] + df['2ndFlrSF'] + df['TotalBsmtSF']
  df['TotalBath'] = df['FullBath'] + 0.5 * df['HalfBath'] + df['BsmtFullBath'] + 0.5 * df['BsmtHalfBath']
  df['HouseAge'] = df['YrSold'] - df['YearBuilt']
```

```python
# 순서형 변수 인코딩
qual_dict = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'None': 0}
ordinals = ['ExterQual','ExterCond','BsmtQual','BsmtCond','HeatingQC','KitchenQual','FireplaceQu','GarageQual','GarageCond']
for col in ordinals:
  train[col] = train[col].map(qual_dict)
  test[col] = test[col].map(qual_dict)
```

```python
# 원핫 인코딩
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

### 모델 설정 및 학습

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import (
    ExtraTreesRegressor,
    GradientBoostingRegressor,
    HistGradientBoostingRegressor,
)
from sklearn.pipeline import Pipeline
```

```python
def rmse(y, pred):
    return np.sqrt(mean_squared_error(y, pred))

# 환경에 따라 없을 수 있어 try/except 처리
try:
    from xgboost import XGBRegressor
    HAS_XGB = True
except Exception:
    HAS_XGB = False

try:
    from lightgbm import LGBMRegressor
    HAS_LGBM = True
except Exception:
    HAS_LGBM = False
```

```python
drop_cols = [c for c in ["SalePrice", "SalePriceLog", "Id"] if c in train.columns]
feature_cols = [c for c in train.columns if c not in drop_cols]

# test 원핫 결과가 train과 다를 수 있으므로 열 동기화
test = test.reindex(columns=feature_cols, fill_value=0)
```

```python
X_df   = train[feature_cols]          # DataFrame (트리계열에 사용)
X_test_df = test[feature_cols]        # DataFrame
y_log = train["SalePriceLog"].values  # 학습용(로그)
y_true = train["SalePrice"].values    # 평가용(원 스케일)
```

```python
models = {}

# 트리/부스팅 (스케일 불필요)
models["extra"] = ExtraTreesRegressor(
    n_estimators=1200, max_depth=None, min_samples_leaf=1,
    n_jobs=-1, random_state=42
)
models["gbr"] = GradientBoostingRegressor(
    n_estimators=3000, learning_rate=0.03, max_depth=3,
    random_state=42
)
models["hgb"] = HistGradientBoostingRegressor(
    learning_rate=0.06, max_depth=4, max_iter=800,
    random_state=42
)

if HAS_XGB:
    models["xgb"] = XGBRegressor(
        n_estimators=2000, learning_rate=0.03, max_depth=4,
        subsample=0.8, colsample_bytree=0.7, reg_lambda=1.0,
        tree_method="hist", n_jobs=-1, random_state=42
    )
if HAS_LGBM:
    models["lgbm"] = LGBMRegressor(
        n_estimators=4000, learning_rate=0.03, num_leaves=31,
        subsample=0.8, colsample_bytree=0.7, n_jobs=-1, random_state=42
    )
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = {}
oof_preds = {}
test_preds = {}

for name, model in models.items():
    oof = np.zeros(len(X_df))
    test_fold_preds = []

    for tr_idx, va_idx in kf.split(X_df):
        X_tr, X_va = X_df.iloc[tr_idx], X_df.iloc[va_idx]
        y_tr, y_va = y_log[tr_idx], y_log[va_idx]

        # 파이프라인이면 fit(X,y)로 알아서 스케일 → 모델
        model.fit(X_tr, y_tr)
        va_pred = np.expm1(model.predict(X_va))
        oof[va_idx] = va_pred

        te_pred = np.expm1(model.predict(X_test_df))
        test_fold_preds.append(te_pred)

    score = rmse(y_true, oof)
    cv_scores[name] = score
    oof_preds[name] = oof
    test_preds[name] = np.mean(test_fold_preds, axis=0)
    print(f"[CV] {name:5s} RMSE: {score:,.1f}")
```

```python
# 성능기반 가중 보팅 (Top-k 모델, weight = 1/RMSE^2)

k = 3  # 상위 몇 개를 합칠지
sorted_models = sorted(cv_scores.items(), key=lambda x: x[1])
top_models = [m for m, _ in sorted_models[:k]]

weights = {m: 1.0 / (cv_scores[m]**2) for m in top_models}
w_sum = sum(weights.values())
weights = {m: w / w_sum for m, w in weights.items()}

oof_blend = np.zeros(len(X_df))
test_blend = np.zeros(len(X_test_df))
for m in top_models:
    oof_blend  += weights[m] * oof_preds[m]
    test_blend += weights[m] * test_preds[m]

blend_score = rmse(y_true, oof_blend)
print(f"[CV] W-VOTE({'+'.join(top_models)}) RMSE: {blend_score:,.1f}")
print("Weights:", {m: round(w, 3) for m, w in weights.items()})
```

```python
# 간단 스태킹 (Level-1: 상위 모델들 고정, Level-2: Ridge 메타)
#  - OOF로 메타 입력 구성 → Ridge(로지타깃) 학습

from sklearn.linear_model import Ridge

stack_base = top_models  # 보통 상위 2~4개 사용
# Level-1 OOF 메타 피처 (로그스케일로 변환해서 안정화)
meta_oof = np.column_stack([np.log1p(oof_preds[m]) for m in stack_base])
meta_test = np.column_stack([np.log1p(test_preds[m]) for m in stack_base])

meta = Ridge(alpha=1.0, random_state=42)

# 내부 5-Fold로 메타 OOF 점검 (선택)
oof_meta = np.zeros(len(X_df))
for tr_idx, va_idx in kf.split(meta_oof):
    meta.fit(meta_oof[tr_idx], y_log[tr_idx])
    oof_meta[va_idx] = np.expm1(meta.predict(meta_oof[va_idx]))

stack_score = rmse(y_true, oof_meta)
print(f"[CV] STACK({'+'.join(stack_base)} -> ridge) RMSE: {stack_score:,.1f}")

# 최종 학습 및 테스트 예측
meta.fit(meta_oof, y_log)
stack_test_pred = np.expm1(meta.predict(meta_test))
```

### 테스트 데이터 예측 및 제출

```python
test_raw = pd.read_csv("./test.csv")

final_pred = stack_test_pred if stack_score <= blend_score else test_blend
final_tag  = "stack" if stack_score <= blend_score else "wvote"

submission = pd.DataFrame({"Id": test_raw["Id"], "SalePrice": final_pred})
submission.to_csv(f"submission_{final_tag}.csv", index=False)
print(f"Saved: submission_{final_tag}.csv")
```
